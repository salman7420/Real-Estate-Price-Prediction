```python
def convert_to_square_yard(row):
    marla_to_sq_Yd = 30.25
    kanal_to_sq_Yd = 20 * marla_to_sq_Yd

    if row['Area Type (Marla, Kanal, Sq. Yd)'] == 'Kanal':
        return row['Size (Marla, Kanal, Sq. Yd)'] * kanal_to_sq_Yd
    elif row['Area Type (Marla, Kanal, Sq. Yd)'] == 'Marla':
        return row['Size (Marla, Kanal, Sq. Yd)'] * marla_to_sq_Yd
    else:
        return row['Size (Marla, Kanal, Sq. Yd)']


df['Area_in_Sq_Yd'] = df.apply(convert_to_square_yard, axis=1)
df
```

#One Hot encoder#

```python
# Converting property_type to categories
one_hot = pd.get_dummies(df['Property Type'])

# Concatenate the original DataFrame and the one-hot encoded DataFrame along the column axis
df = pd.concat([df, one_hot], axis=1)

df = df.drop("Property Type", axis=1)


# Converting locations to categories
one_hot = pd.get_dummies(df['Location'])

# Concatenate the original DataFrame and the one-hot encoded DataFrame along the column axis
df = pd.concat([df, one_hot], axis=1)

df = df.drop("Location", axis=1)


# Converting city to categories
one_hot = pd.get_dummies(df['City'])

# Concatenate the original DataFrame and the one-hot encoded DataFrame along the column axis
df = pd.concat([df, one_hot], axis=1)

df = df.drop("City", axis=1)



# Converting purpose to categories
one_hot = pd.get_dummies(df['Province'])

# Concatenate the original DataFrame and the one-hot encoded DataFrame along the column axis
df = pd.concat([df, one_hot], axis=1)

df = df.drop('Province', axis=1)
df
```

## Applying Regression algorithm

```python
from sklearn.model_selection import train_test_split

X = df.drop("price_millions", axis = 1)
y = df["price_millions"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=15)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
mean_squared_error_array = []
r2_score_array = []
algorithm = []
headings = ['algorithm', 'meanSquare', 'r2score']
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score, mean_squared_error
dtr = DecisionTreeRegressor()
dtr.fit(X_train, y_train)
y_preds = dtr.predict(X_test)
algorithm.append("DecisionTreeRegressor")
m_s_e = mean_squared_error(y_test, y_preds)
mean_squared_error_array.append(m_s_e)
m_s_e
```

```python
r2 = r2_score(y_test, y_preds)
r2_score_array.append(r2)
r2
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor()
rfr.fit(X_train, y_train)
y_preds = rfr.predict(X_test)
algorithm.append("RandomForestRegressor")
m_s_e = mean_squared_error(y_test, y_preds)
mean_squared_error_array.append(m_s_e)
m_s_e
```

```python
r2 = r2_score(y_test, y_preds)
r2_score_array.append(r2)
r2
```

```python
from sklearn.svm import SVR
svr = SVR()
svr.fit(X_train, y_train)
y_preds = svr.predict(X_test)
algorithm.append("SVR")
m_s_e = mean_squared_error(y_test, y_preds)
mean_squared_error_array.append(m_s_e)
m_s_e
```

```python
r2 = r2_score(y_test, y_preds)
r2_score_array.append(r2)
r2
```

```python
plt.xlabel('algorithms')
plt.ylabel('mse')
# plt.legend()
sns.barplot(data= df, x = algorithm, y = mean_squared_error_array)
```

```python
for i in range(len(r2_score_array)):
  r2_score_array[i] = r2_score_array[i]*100

plt.xlabel('algorithms')
plt.ylabel('r2 score %')
sns.barplot(data= df, x = algorithm, y = r2_score_array)
```
